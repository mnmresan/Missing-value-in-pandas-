```python
import pandas as pd
import numpy as np
dataframe = {'day':['1/1/2017','1/4/2017','1/5/2017','1/6/2017','1/7/2017','1/8/2017','1/9/2017','1/10/2017','1/11/2017'],
             'temperature':['32',"NaN",'28','nan','32','np','np','34','40'],
             'windspeed':['6','9','NULL','7','NULL','','NULL','8','12'],
             'event':['Rain','Sunny','Snow','NULL','Rain','Sunny','NULL','Cloudy','Sunny']}
df= pd.DataFrame(dataframe)
df = df.reset_index(drop=True)
df2 = pd.read_csv('weather_data.csv')
df
```

Fill all the missing value with Null Values using pandas and numply

```python
df =df.replace(['NaN','np','NULL','nan',""],np.nan)
# df[['temperature','windspeed','event']]=df[['temperature','windspeed','event']].replace(['NaN','np','NULL'],np.nan)
df.isna()
```

```python
df
```

Fill all the null value with 0

```python
df.fillna(0)
```

Fill the each column with different values 

```python
df.fillna({'temperature':0,
           'windspeed':10,
           'event': "no event"})
```

fill the null value with forward filling method

```python
new_df = df.fillna(method='ffill')
new_df
```

Fill the null with backward filling method

```python
new_df = df.fillna(method='bfill')
new_df
```

Using the limit funtion work with bfill and ffill

```python
new_df = df.fillna(method='ffill',limit=1)
new_df 
```

Filling the missing values horizontaly

```python
new_df = df.fillna(method='bfill',limit=1,axis='columns')
new_df 
```
